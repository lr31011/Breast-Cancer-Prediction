```python
#Import the Requried Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
#We will take the inbuilt brest-cancer data from sklearn
from sklearn.datasets import load_breast_cancer
```

```python
#load the data
cancer = load_breast_cancer()
```

```python
#input variables
inputs = cancer.data
inputs
```

```python
#Ouput/target variable
output = cancer.target
output
```

```python
# 0 - 'malignant' , 1 - 'benign'
cancer.target_names
```

```python
#Dataset Description
print(cancer.DESCR)
```

```python
#lets Create a dataframe
df = pd.DataFrame(inputs,columns = cancer.feature_names)
df['Output'] = output
df.head()
```

```python
np.c_[cancer.data,cancer.target]
```

```python
np.append(cancer.feature_names,'target')
```

```python
#Another way to combine inputs and outputs
df = pd.DataFrame(np.c_[cancer.data,cancer.target],columns = np.append(cancer.feature_names,'Output'))
df.head()
```

```python
cancer.feature_names
```

```python
#Data Visualisations
#First we will plot pairplot(A pairs plot allows us to see both distribution of single variables and relationships between two variables .)
sns.pairplot(df,vars = ['mean radius', 'mean texture', 'mean perimeter', 'mean area',
       'mean smoothness', 'mean compactness'],hue = 'Output')
# Strong correlation between the mean radius and mean perimeter,mean radius and mean area, mean area and mean primeter
```

```python
# Countplot to obtain the No of Unique Values in a data
sns.countplot(df['Output'])
```

```python
#Scatter plot between two vairables
sns.scatterplot(x = 'mean area', y = 'mean smoothness', hue = 'Output', data = df)
```

```python
df.corr()
```

```python
#Let's check the correlations between the input variables 
#for that we will use heatmap
plt.figure(figsize = (20,20))
sns.heatmap(df.corr(),annot = True)
```

```python
#Feature Scaling
X = df.drop(['Output'],axis = 1)
```

```python
X.head()
```

```python
Y = df.loc[:,'Output']
```

```python
type(Y)
```

```python
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
X = scale.fit_transform(X)
```

```python
#Split the data for training and testing the model
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,stratify = Y,test_size = 0.3,random_state = 42)
```

```python
X_train.shape,X_test.shape
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
logit = LogisticRegression()
logit.fit(X_train,Y_train)
```

```python
logit.coef_
```

```python
logit.intercept_
```

```python
test_prediction = logit.predict(X_test)
test_prediction
```

```python
from sklearn import metrics
print(metrics.classification_report(test_prediction,Y_test))
```

```python
#Accuracy Score
metrics.accuracy_score(test_prediction,Y_test)
```

```python
metrics.confusion_matrix(test_prediction,Y_test)
```

```python
sns.heatmap(metrics.confusion_matrix(test_prediction,Y_test),annot = True,cmap = 'viridis')
```

```python
#fine tune the parameters
from sklearn.model_selection import GridSearchCV
parameters = {'C' : [0.001,0.01,0.1,1,10,100]}
grid_search = GridSearchCV(param_grid = parameters , estimator = logit  , cv = 5 , scoring = 'accuracy',n_jobs = -1)
```

```python
grid_search.fit(X_train,Y_train)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_params_
```

```python
logit = LogisticRegression(C = 10)
logit.fit(X_train,Y_train)
```

```python
test_prediction = logit.predict(X_test)
test_prediction
```

```python
print(metrics.classification_report(test_prediction,Y_test))
```

```python
#Accuracy Score
metrics.accuracy_score(test_prediction,Y_test)
```

```python
metrics.confusion_matrix(test_prediction,Y_test)
```

```python
sns.heatmap(metrics.confusion_matrix(test_prediction,Y_test),annot = True,cmap = 'viridis')
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train,Y_train)
```

```python
dt_predictions = dt.predict(X_test)
print(metrics.classification_report(dt_predictions,Y_test))
```

```python
metrics.accuracy_score(dt_predictions,Y_test)
```

```python
sns.heatmap(metrics.confusion_matrix(dt_predictions,Y_test),annot = True)
```

```python
#Tune the parameters
# max_depth = Depth of the tree(Higher the value leads to OverFitting)
# min_samples_leaf = minimum samples at the leaf(Higher the value leads to UnderFitting)
# min_samples_split = minimun samples required to split the tree(Higher the value leads to UnderFitting)
parameters = {'max_depth' : list(range(1,10,2)) , 'min_samples_leaf' : [0.05,0.1,0.15,0.2,0.25,0.3],'min_samples_split' : [0.05,0.1,0.15,0.2,0.25,0.3]}
grid_search = GridSearchCV(param_grid = parameters,estimator = dt,cv = 5,scoring = 'accuracy')
```

```python
grid_search.fit(X_train,Y_train)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
dt = DecisionTreeClassifier(max_depth = 3, min_samples_leaf = 0.05 , min_samples_split = 0.05)
dt.fit(X_train,Y_train)
dt_prediction = dt.predict(X_test)
print(metrics.classification_report(dt_prediction,Y_test))
```

```python
metrics.accuracy_score(dt_prediction,Y_test)
```
